```python
import pandas as pd

df=pd.read_csv("Books_Data.csv", sep=";")

df['operatingCost']=df['gross sales']-df['publisher revenue']
df.drop(columns=['sales rank'], inplace=True)

print(df.head())
```

```python

mean_gross_sales = df['gross sales'].mean()
median_gross_sales = df['gross sales'].median()
std_dev_gross_sales = df['gross sales'].std()

print("Mean gross sales:", mean_gross_sales)
print("Median gross sales:", median_gross_sales)
print("Standard deviation of gross sales:", std_dev_gross_sales)
```

```python
df['Publishing Year']=df['Publishing Year'].astype(str).str.split('.').str[0]

incorrect_dates_index=df[df['Publishing Year'].str.contains('[0-9]')].index
df.drop(index=incorrect_dates_index, inplace=True)

print(df.head())
```
